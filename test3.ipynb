```python
df1=pd.read_csv("../input/newsdata/Fake.csv")
df2=pd.read_csv("../input/newsdata/True.csv")
```

```python
df1['result']=0
df2['result']=1
```

```python
df=pd.concat([df1,df2],axis=0)
```

```python
df.head()
```

```python
df = df.drop(["title", "subject","date"], axis = 1)
```

```python
def fun(x):
    x=x.lower()
    x=re.sub('\n', '',x)
    x=re.sub('\[.*?\]', '',x)
    x=re.sub('\w*\d\w*', '',x) 
    x=re.sub("\\W"," ",x)
    x=re.sub('<.*?>+', '',x)
    x=re.sub('[%s]' % re.escape(string.punctuation), '',x)
    x=re.sub('https?://\S+|www\.\S+', '',x)
    return x
```

```python
df['text'] = df['text'].apply(fun)
```

```python
x = df["text"]
y = df["result"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
v= TfidfVectorizer()
xtrain = v.fit_transform(x_train)
xtest = v.transform(x_test)
```

```python
logreg= LogisticRegression()
logreg.fit(xtrain,y_train)
ypred=logreg.predict(xtest)
```

```python
logreg.score(xtest,y_test)
```

```python
print(classification_report(y_test,ypred))
```

```python
randomforest= RandomForestClassifier(random_state=0)
randomforest.fit(xtrain, y_train)
ypred=randomforest.predict(xtest)
print(randomforest.score(xtest, y_test))
print(classification_report(y_test,ypred))
```

```python
dectree= DecisionTreeClassifier()
dectree.fit(xtrain, y_train)
```

```python
ypred =dectree.predict(xtest)
print(dectree.score(xtest, y_test))
print(classification_report(y_test,ypred))
```
